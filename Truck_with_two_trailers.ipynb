```python
import numpy as np
from numpy import array
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import patches
from matplotlib import transforms
%matplotlib inline
from numpy.linalg import norm
```

```python
def visualize_combination(truck_translation,\
                          truck_rotation,\
                          dolly_rotation,\
                          trailer_rotation,\
                          steering_percentage,\
                          destination_translation,\
                          destination_rotation):
    
    yard_shape = array([30,10])
    truck_shape = array([7,2])
    cab_shape = array([1.5,2])
    destination_shape = array([9,2])
    dolly_shape = array([2,2])
    trailer_shape = array([9,2])
    shaft_shape = array([2,0.2])
    wheel_shape = array([0.8,0.4])
    drive_wheel_shape = array([0.8,0.6])
    hitch_radius = 0.1
    
    item_translations_truck = array([
        [0,0],
        [3.75,0],
        [-1.5,0],
        [0,-0.7],
        [0,0.7],
        [-1,-0.8],
        [-1,0.8],
        [3.5,-0.8],
        [3.5,0.8]])
    item_translations_dolly = array([
        [0,0],
        [1.5,0],
        [-0.5,0],
        [-1,-0.8],
        [-1,0.8],
        [0,-0.8],
        [0,0.8]])
    item_translations_trailer = array([
        [0,0],
        [-1,-0.8],
        [-1,0.8],
        [0,-0.8],
        [0,0.8],
        [1,-0.8],
        [1,0.8]])
    
    rotation_center_truck = array([2.5,1])
    rotation_center_dolly = array([1.5,1])
    rotation_center_trailer = array([2.5,1])
    
    hitch_translation_truck = -1.5
    hitch_translation_dolly_truck = 2.5
    hitch_translation_dolly_trailer = -0.5
    hitch_translation_trailer = 5.5
    
    maximal_steering_angle = 60
    
    # truck item numbers
    # 0: truck
    # 1: cab
    # 2: hitch
    # 3: wheel middle right
    # 4: wheel middle left
    # 5: wheel back right
    # 6: wheel back left
    # 7: wheel front right
    # 8: wheel front left
    
    # dolly item numbers
    # 0: chassis
    # 1: shaft
    # 2: hitch
    # 3: wheel back right
    # 4: wheel back left
    # 5: wheel front right
    # 6: wheel front left
    
    # trailer item numbers
    # 0: chassis
    # 1: shaft
    # 2: wheel back right
    # 3: wheel back left
    # 4: wheel front right
    # 5: wheel front left
    
    item_steering_rotations_truck = array([0,0,0,0,0,-0.6,-0.6,1,1])
    item_steering_rotations_dolly = array([0,0,0,-0.4,-0.4,0,0])
    item_steering_rotations_trailer = array([0,-0.2,-0.2,0,0,0.2,0.2])
    
    ############################################################################################################################
    
    fig = plt.figure(figsize=(15,15))
    ax1 = fig.add_subplot(111, aspect='equal')
    ax1.axes.get_xaxis().set_visible(False)
    ax1.axes.get_yaxis().set_visible(False)
    
    ## Truck plotting
    wheel_rectangles = []
    everything = []
    truck_rectangle = patches.Rectangle(-rotation_center_truck,\
                                        truck_shape[0],\
                                        truck_shape[1],\
                                        color="black",\
                                        alpha=0.50)
    cab_rectangle = patches.Rectangle(-cab_shape/2,\
                                      cab_shape[0],\
                                      cab_shape[1],\
                                      color="green",\
                                      alpha=0.50)
    drive_wheel_rectangles = [patches.Rectangle(-drive_wheel_shape/2,\
                                          drive_wheel_shape[0],\
                                          drive_wheel_shape[1],\
                                          color="black",\
                                          alpha=0.50)\
                                          for drive_wheel_number in range(2)]
    wheel_rectangles = [patches.Rectangle(-wheel_shape/2,\
                                          wheel_shape[0],\
                                          wheel_shape[1],\
                                          color="black",\
                                          alpha=0.50)\
                                          for wheel_number in range(5)]
    hitch_circle = patches.Circle((0, 0),\
                                  hitch_radius,\
                                  color="black",\
                                  alpha=0.50)
    everything = [truck_rectangle]+[cab_rectangle]+[hitch_circle]+drive_wheel_rectangles+wheel_rectangles
    
    item_number = 0
    item_rotation = []
    item_translation = []
    for item_number in range(len(everything)-1):
        # rotate wheels
        item_rotation = item_rotation\
                        +[transforms.Affine2D().rotate_deg(steering_percentage\
                                                           *maximal_steering_angle\
                                                           *item_steering_rotations_truck[item_number])]
        # translate wheels
        item_translation = item_translation\
                           +[transforms.Affine2D().translate(item_translations_truck[item_number,0],\
                                                             item_translations_truck[item_number,1])]
    # rotate truck with wheels
    rotation_everything = transforms.Affine2D().rotate_deg(truck_rotation)
    # translate truck with wheels
    translation_everything = transforms.Affine2D().translate(truck_translation[0],\
                                                             truck_translation[1])
    
    for item_number in range(len(everything)-1):
        item_transformation = item_rotation[item_number]\
                              +item_translation[item_number]\
                              +rotation_everything\
                              +translation_everything\
                              +ax1.transData
        everything[item_number].set_transform(item_transformation)
        ax1.add_patch(everything[item_number])
        
    ## Destination plotting
    destination_rectangle = patches.Rectangle(-destination_shape/2,\
                                              destination_shape[0],\
                                              destination_shape[1],\
                                              color="red",\
                                              alpha=0.50)
    rotation_destination = transforms.Affine2D().rotate_deg(destination_rotation)
    translation_destination = transforms.Affine2D().translate(destination_translation[0],\
                                                              destination_translation[1])
    destination_transformation = rotation_destination\
                                 +translation_destination\
                                 +ax1.transData
    destination_rectangle.set_transform(destination_transformation)
    ax1.add_patch(destination_rectangle)
    
    ## Dolly plotting
    wheel_rectangles = []
    everything = []
    dolly_rectangle = patches.Rectangle(-rotation_center_dolly,\
                                        dolly_shape[0],\
                                        dolly_shape[1],\
                                        color="black",\
                                        alpha=0.50)
    shaft_rectangle = patches.Rectangle(-shaft_shape/2,\
                                        shaft_shape[0],\
                                        shaft_shape[1],\
                                        color="black",\
                                        alpha=0.50)
    wheel_rectangles = [patches.Rectangle(-wheel_shape/2,\
                                          wheel_shape[0],\
                                          wheel_shape[1],\
                                          color="black",\
                                          alpha=0.50)\
                                          for wheel_number in range(5)]
    hitch_circle = patches.Circle((0, 0),\
                                  hitch_radius,\
                                  color="black",\
                                  alpha=0.50)
    everything = [dolly_rectangle]+[shaft_rectangle]+[hitch_circle]+wheel_rectangles
    
    item_number = 0
    item_rotation = []
    item_translation = []
    for item_number in range(len(everything)-1):
        # rotate wheels
        item_rotation = item_rotation\
                        +[transforms.Affine2D().rotate_deg(-dolly_rotation\
                                                           *item_steering_rotations_dolly[item_number])]
        # translate wheels
        item_translation = item_translation\
                           +[transforms.Affine2D().translate(item_translations_dolly[item_number,0],\
                                                             item_translations_dolly[item_number,1])]
    # rotate dolly with wheels
    rotation_everything = transforms.Affine2D().rotate_deg(truck_rotation\
                                                           +dolly_rotation)
    # translate dolly with wheels
    translation_everything = transforms.Affine2D().translate(truck_translation[0]\
                                                             +hitch_translation_truck\
                                                             *np.cos(np.deg2rad(truck_rotation))\
                                                             -hitch_translation_dolly_truck\
                                                             *np.cos(np.deg2rad(truck_rotation+dolly_rotation)),\
                                                             truck_translation[1]\
                                                             +hitch_translation_truck\
                                                             *np.sin(np.deg2rad(truck_rotation))\
                                                             -hitch_translation_dolly_truck\
                                                             *np.sin(np.deg2rad(truck_rotation+dolly_rotation)))
    
    for item_number in range(len(everything)-1):
        item_transformation = item_rotation[item_number]\
                              +item_translation[item_number]\
                              +rotation_everything\
                              +translation_everything\
                              +ax1.transData
        everything[item_number].set_transform(item_transformation)
        ax1.add_patch(everything[item_number])
        
    ## Trailer plotting
    wheel_rectangles = []
    everything = []
    trailer_rectangle = patches.Rectangle(-rotation_center_trailer,\
                                          trailer_shape[0],\
                                          trailer_shape[1],\
                                          color="black",\
                                          alpha=0.50)
    wheel_rectangles = [patches.Rectangle(-wheel_shape/2,\
                                          wheel_shape[0],\
                                          wheel_shape[1],\
                                          color="black",\
                                          alpha=0.50)\
                                          for wheel_number in range(7)]
    everything = [trailer_rectangle]+wheel_rectangles
    
    item_number = 0
    item_rotation = []
    item_translation = []
    for item_number in range(len(everything)-1):
        # rotate wheels
        item_rotation = item_rotation\
                        +[transforms.Affine2D().rotate_deg(-trailer_rotation\
                                                           *item_steering_rotations_trailer[item_number])]
        # translate wheels
        item_translation = item_translation\
                           +[transforms.Affine2D().translate(item_translations_trailer[item_number,0],\
                                                             item_translations_trailer[item_number,1])]
    # rotate trailer with wheels
    rotation_everything = transforms.Affine2D().rotate_deg(truck_rotation\
                                                           +dolly_rotation\
                                                           +trailer_rotation)
    # translate trailer with wheels
    translation_everything = transforms.Affine2D().translate(truck_translation[0]\
                                                             +hitch_translation_truck\
                                                             *np.cos(np.deg2rad(truck_rotation))\
                                                             +(-hitch_translation_dolly_truck+hitch_translation_dolly_trailer)\
                                                             *np.cos(np.deg2rad(truck_rotation+dolly_rotation))\
                                                             -hitch_translation_trailer\
                                                             *np.cos(np.deg2rad(truck_rotation+dolly_rotation+trailer_rotation)),\
                                                             truck_translation[1]\
                                                             +hitch_translation_truck\
                                                             *np.sin(np.deg2rad(truck_rotation))\
                                                             +(-hitch_translation_dolly_truck+hitch_translation_dolly_trailer)\
                                                             *np.sin(np.deg2rad(truck_rotation+dolly_rotation))\
                                                             -hitch_translation_trailer\
                                                             *np.sin(np.deg2rad(truck_rotation+dolly_rotation+trailer_rotation)))
    
    for item_number in range(len(everything)-1):
        item_transformation = item_rotation[item_number]\
                                            +item_translation[item_number]\
                                            +rotation_everything\
                                            +translation_everything\
                                            +ax1.transData
        everything[item_number].set_transform(item_transformation)
        ax1.add_patch(everything[item_number])
    
    
    plt.xlim(0, yard_shape[0])
    plt.ylim(0, yard_shape[1])
    plt.show()
```

```python
truck_translation = array([8,5])
truck_rotation = 160
dolly_rotation = 10
trailer_rotation = 20
steering_percentage = 0.5
destination_translation = array([25.5,5])
destination_rotation = 180


############################################################################################################################

visualize_combination(truck_translation,\
                      truck_rotation,\
                      dolly_rotation,\
                      trailer_rotation,\
                      steering_percentage,\
                      destination_translation,\
                      destination_rotation)
```

```python
def vector_rotation(vector,angle):
    rotation_matrix = array([[np.cos(np.deg2rad(angle)),-np.sin(np.deg2rad(angle))],
                             [np.sin(np.deg2rad(angle)),np.cos(np.deg2rad(angle))]])
    vector = np.dot(rotation_matrix,vector)
    return vector

def constant_rotation(constant,angle):
    rotation_vector = array([np.cos(np.deg2rad(angle)),\
                             np.sin(np.deg2rad(angle))])
    vector = rotation_vector*constant
    return vector

def endpoint_movement(vector,angle):
    movement = vector_rotation(vector,angle)-vector
    return movement

def angle_two_vectors(vector1,vector2):
    if norm(vector1) != 0 and norm(vector1) != 0:
        angle = np.rad2deg(np.arccos(np.dot(vector1,vector2)/(norm(vector1)*norm(vector2))))
    else:
        angle = 0
    return angle
```

```python
class Simulate_combination():
    def __init__(self,destination_translation,destination_rotation):
        
        self.truck_translation = array([np.float(18),np.float(5)])
        self.truck_rotation = 0
        self.dolly_rotation = 20
        self.trailer_rotation = 20
        
        self.wheelbase_truck = 3.5
        self.maximal_steering_angle = 60
        self.hitch_translation_truck = -1.5
        self.hitch_translation_dolly_truck = 2.5
        self.hitch_translation_dolly_trailer = -0.5
        self.hitch_translation_trailer = 5.5
        
        self.step_size = 1e-1
        self.plotting_interval = 1e1
        
        self.step_number = 1
               
        self.destination_rotation = destination_rotation
        self.destination_translation = destination_translation
    
    def run(self,velocity,steering_percentage):
        
        steering_angle = steering_percentage*self.maximal_steering_angle
        distance = velocity*self.step_size
        
        truck_movement = constant_rotation(distance,self.truck_rotation)
        dolly_movement = vector_rotation(truck_movement,self.dolly_rotation)
        trailer_movement = vector_rotation(dolly_movement,self.trailer_rotation)
        
        hitch_vector_truck = constant_rotation(self.hitch_translation_truck,\
                                               self.truck_rotation)
        hitch_vector_dolly_truck = constant_rotation(self.hitch_translation_dolly_truck,\
                                                     self.truck_rotation+\
                                                     self.dolly_rotation)
        hitch_vector_dolly_trailer = constant_rotation(self.hitch_translation_dolly_trailer,\
                                                       self.truck_rotation+\
                                                       self.dolly_rotation)
        hitch_vector_trailer = constant_rotation(self.hitch_translation_trailer,\
                                                 self.truck_rotation+\
                                                 self.dolly_rotation+\
                                                 self.trailer_rotation)
        
        if norm(truck_movement) != 0:
            
            old_truck_rotation = self.truck_rotation
            
            ## Truck movement
            if steering_percentage != 0:
                lock = self.wheelbase_truck/np.tan(np.deg2rad(steering_angle))
                rotation_truck = np.rad2deg(np.arcsin(distance/lock))/2
            self.truck_rotation += rotation_truck
            
            translation_truck = vector_rotation(truck_movement,rotation_truck)
            self.truck_translation += translation_truck
          
            step_rotation_truck = self.truck_rotation-old_truck_rotation
            old_dolly_rotation = self.dolly_rotation
            
            ## Dolly movement
            truck_hitch_movement = truck_movement\
                                   +endpoint_movement(hitch_vector_truck,\
                                                      step_rotation_truck)
            
            dolly_movement = vector_rotation(truck_movement,self.dolly_rotation)
            
            rotation_dolly = np.sign(self.dolly_rotation)\
                                     *angle_two_vectors(hitch_vector_dolly_truck\
                                     -dolly_movement\
                                     +truck_hitch_movement,\
                                     truck_movement)
            self.dolly_rotation = -step_rotation_truck+rotation_dolly
            
            step_rotation_dolly = self.dolly_rotation-old_dolly_rotation
            
            ## Trailer movement
            dolly_hitch_movement = dolly_movement\
                                   +endpoint_movement(hitch_vector_dolly_trailer,\
                                                      step_rotation_dolly)
            
            trailer_movement = vector_rotation(dolly_movement,self.trailer_rotation)
            
            rotation_trailer = np.sign(self.trailer_rotation)\
                                       *angle_two_vectors(hitch_vector_trailer\
                                       -trailer_movement\
                                       +dolly_hitch_movement,\
                                       dolly_movement)
            self.trailer_rotation = -step_rotation_truck-step_rotation_dolly+rotation_trailer
                
        if np.mod(self.step_number,self.plotting_interval) == 0:
            self.step_number = 1
            visualize_combination(self.truck_translation,\
                                  self.truck_rotation,\
                                  self.dolly_rotation,\
                                  self.trailer_rotation,\
                                  steering_percentage,\
                                  self.destination_translation,\
                                  self.destination_rotation)
        else:
            self.step_number += 1
            
        return self.truck_translation,self.truck_rotation,self.dolly_rotation,self.trailer_rotation
```

```python
destination_translation = array([4.5,5])
destination_rotation = 180

simulate_combination = Simulate_combination(destination_translation,destination_rotation)

for step_number in range(np.int(1e2)):
    velocity = np.sin(step_number/1e2*np.pi)
    steering_percentage = np.sin(step_number/1e2*np.pi)
    simulate_combination.run(velocity,steering_percentage)
```
